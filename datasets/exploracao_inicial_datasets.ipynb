# Exploração Inicial Datasets

Notebook criado para apresentar, para os datasets dessa pasta, quais são as colunas (e seus tipos), e uma amostra dos registros dos mesmos.

Os datasets são:
- [`WA_Fn-UseC_-HR-Employee-Attrition`](https://www.kaggle.com/pavansubhasht/ibm-hr-analytics-attrition-dataset)
- [`Nutrition Facts for McDonald's Menu`](https://www.kaggle.com/mcdonalds/nutrition-facts)
- [`Spotify Song Attributes`](https://www.kaggle.com/geomack/spotifyclassification)
- [`Chocolate Bar Ratings`](https://www.kaggle.com/rtatman/chocolate-bar-ratings)
- [`Uber Drives`](https://www.kaggle.com/zusmani/uberdrives)
- [`New York City Airbnb Open Data`](https://www.kaggle.com/dgomonov/new-york-city-airbnb-open-data)

Para isto, vamos usar os seguintes métodos:

- `df.sample()`:
> Permite selecionar valores aleatoriamente de uma série ou DataFrame. É útil quando queremos selecionar uma amostra aleatória de uma distribuição.

- `df.describe()`:
> Calcula estatísticas básicas para colunas numéricas que são contagem, média, desvio padrão, valores mínimo e máximo, mediana, primeiro e terceiro quartil. Assim, ele fornece um resumo estatístico do quadro de dados.

- `df.info()`:
> Apresenta quais são as colunas, os tipos das mesmas, e quantos registros delas são não nulos.
